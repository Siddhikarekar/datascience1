```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
#reading csv file
df=pd.read_csv('bank-full (1).csv')
```

```python
#to display first 5 rows
df.head()
```

```python
target= df[df.Target==1]
target.shape
```

```python
# converting the string values into integer 
data1=pd.get_dummies(df,columns=['job','marital','education','contact','poutcome','month'])
data1
```

```python
data1.info()
```

```python
#To display columns
pd.set_option("display.max.columns", None)
data1
```

```python
#for converting the int data types into string
data1[['default','housing','loan','Target']].astype(str)
```

```python
# Custom Binary Encoding of Binary o/p variables
data1['default'] = np.where(data1['default'].str.contains("yes"), 1, 0)
data1['housing'] = np.where(data1['housing'].str.contains("yes"), 1, 0)
data1['loan'] = np.where(data1['loan'].str.contains("yes"), 1, 0)
data1['Target'] = np.where(data1['Target'].str.contains("yes"), 1, 0)
data1
```

```python
data1.info()
```

```python
## Dividing our data into input and output variables
x=pd.concat([data1.iloc[:,0:10],data1.iloc[:,11:]],axis=1)
y=data1.iloc[:,10]
```

```python
## traning the model 
lr=LogisticRegression()
lr.fit(x,y)
```

```python
#predicting the values for x varaible
y_pred=lr.predict(x)
y_pred
```

```python
y_pred_df=pd.DataFrame({'actual_y':y,'y_pred_prob':y_pred})
y_pred_df
```

```python
# Confusion Matrix for the model accuracy
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix = confusion_matrix(y,y_pred)
confusion_matrix
```

```python
import seaborn as sns
```

```python
sns.heatmap(pd.DataFrame(confusion_matrix),annot=True,cmap='RdBu',fmt='g')
plt.title('Confusion matrix')
plt.ylabel('Actual')
plt.xlabel('predicted')
```

```python
lr.predict_proba(x)[:,1] 
```

```python
# ROC Curve plotting and finding AUC value
fpr,tpr,thresholds=roc_curve(y,lr.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
auc=roc_auc_score(y,y_pred)

plt.plot(fpr,tpr,color='red',label='logit model(area  = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()

print('auc accuracy:',auc)
```
