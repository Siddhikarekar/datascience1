```python
# Converting long data into wide data using pivot table
books_data=books_df.pivot_table(index='userId',columns='Booktitle',values='Bookrating').reset_index(drop=True)
books_data
```

```python
# Replacing the index values by unique UserId
books_data.index=books_df['userId'].unique()
books_data
```

```python
# filling NaNs with 0 values
books_data.fillna(0,inplace=True)
books_data
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( books_data.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = books_df.userId.unique()
user_sim_df.columns = books_df.userId.unique()
```

```python
def get_recommendation(customer_id):
        
        tem=list(user_sim_df.sort_values([customer_id],ascending=False).head(3).index)

        book_list=list(books_df[books_df['userId'].isin(tem)]['Booktitle'])

        recommended_books = set(book_list)-set(books_df[books_df['userId']==customer_id]['Booktitle'])
        
        return recommended_books
```
